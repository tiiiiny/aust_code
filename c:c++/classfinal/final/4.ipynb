```python
import matplotlib.pyplot as plt
import numpy as np

# 创建一个新的图形
fig, ax = plt.subplots()

# 设置坐标轴的范围
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)

# 画圆
d=5
circle = plt.Circle((0, 0), d, color='blue', fill=False)
ax.add_artist(circle)

# 画点
points = [(-3,4), (-6,3), (4, 5), (-5, 3)]
for point in points:
    ax.plot(point[0], point[1], 'ro')  # 'ro' 表示红色的点

# 画坐标轴
ax.axhline(y=0, color='black')
ax.axvline(x=0, color='black')

# 设置比例相等
ax.set_aspect('equal', adjustable='box')

# 显示图形
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# 创建一个新的图形和坐标轴
fig, ax = plt.subplots()

# 设置坐标轴的范围
ax.set_xlim(-15, 10)
ax.set_ylim(-5, 10)

# 定义圆心和半径
centers = [(-3, 4), (-6, 3)]
radius = 5

# 画圆并标出与 x 轴的交点
for center in centers:
    circle = plt.Circle(center, radius, color='blue', fill=False)
    ax.add_artist(circle)

# 画坐标轴
ax.axhline(y=0, color='black')  # 横轴
ax.axvline(x=0, color='black')  # 纵轴

# 设置比例相等
ax.set_aspect('equal', adjustable='box')

# 显示图形
plt.grid(True)  # 添加网格
plt.show()  # 显示图形
```
